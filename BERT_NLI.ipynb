# Datasets

```python
import csv

def tsv2csv (tsv_filepath) :
    with open(tsv_filepath, 'r') as tsv_f:
        tsv = csv.reader(tsv_f, delimiter = '\t')
        csv_filepath = tsv_filepath[:-4] + '.csv'
        with open(csv_filepath, 'w', newline='\n') as csv_f:
            writer = csv.writer(csv_f, delimiter=",")
            for row in tsv:
                writer.writerow(row)
            return csv_filepath
```

```python
#@title JGLUE
import pandas as pd

JGLUE_DIRPATH = '/content/drive/MyDrive/Colab Notebooks/SeniorThesis/DATASETS/JGLUE/datasets/jnli-v1.1/'
JGLUE_FILENAME = { 'train' : 'train-v1.1.json' , 'valid' : 'valid-v1.1.json' }
JGLUE_COLS = [ 'sentence1' , 'sentence2' , 'label' ]
COLUMNS = [ 'sentence1' , 'sentence2' , 'label' ]

jglue_train_df = pd.read_json( JGLUE_DIRPATH + JGLUE_FILENAME['train'] , lines=True , encoding='utf-8' )
jglue_train_df = jglue_train_df.loc[ : , JGLUE_COLS ]
jglue_train_df.columns = COLUMNS
jglue_valid_df = pd.read_json( JGLUE_DIRPATH + JGLUE_FILENAME['valid'] , lines=True , encoding='utf-8' )
jglue_valid_df = jglue_valid_df.loc[ : , JGLUE_COLS ]
jglue_valid_df.columns = COLUMNS

jglue_train_df
```

```python
#@title JSICK
import pandas as pd

JSICK_DIRPATH = '/content/drive/MyDrive/Colab Notebooks/SeniorThesis/DATASETS/JSICK/jsick/'
JSICK_FILENAME = { 'train' : 'train.tsv' , 'valid' : 'test.tsv' }
JSICK_COLS = [ 'sentence_A_Ja' , 'sentence_B_Ja' , 'entailment_label_Ja' ]
COLUMNS = [ 'sentence1' , 'sentence2' , 'label' ]

jsick_train_filepath = tsv2csv(JSICK_DIRPATH + JSICK_FILENAME['train'])
jsick_valid_filepath = tsv2csv(JSICK_DIRPATH + JSICK_FILENAME['valid'])

jsick_train_df = pd.read_csv( jsick_train_filepath , encoding='utf-8' , header=0 )
jsick_train_df = jsick_train_df.loc[ : , JSICK_COLS ]
jsick_train_df.columns = COLUMNS
jsick_valid_df = pd.read_csv( jsick_valid_filepath , encoding='utf-8' , header=0 )
jsick_valid_df = jsick_valid_df.loc[ : , JSICK_COLS ]
jsick_valid_df.columns = COLUMNS

jsick_train_df.index = len(jglue_train_df.index) + jsick_train_df.index
jsick_valid_df.index = len(jglue_valid_df.index) + jsick_valid_df.index

jsick_train_df
```

```python
#@title JNLI KYOTO
import pandas as pd

JNLI_KYOTO_DIRPATH = '/content/drive/MyDrive/Colab Notebooks/SeniorThesis/DATASETS/JSNLI_KYOTO/jsnli_1.1/'
JNLI_KYOTO_FILENAME = { 'train' : 'train_w_filtering.tsv' , 'valid' : 'dev.tsv' }
JNLI_KYOTO_COLUMNS = [ 'label' , 'sentence1' , 'sentence2' ]
COLUMNS = [ 'sentence1' , 'sentence2' , 'label' ]

jnli_kyoto_train_filepath = tsv2csv( JNLI_KYOTO_DIRPATH + JNLI_KYOTO_FILENAME['train'] )
jnli_kyoto_valid_filepath = tsv2csv( JNLI_KYOTO_DIRPATH + JNLI_KYOTO_FILENAME['valid'] )

jnli_kyoto_train_df = pd.read_csv( jnli_kyoto_train_filepath , encoding='utf-8' , header=None, names=JNLI_KYOTO_COLUMNS )
jnli_kyoto_train_df = jnli_kyoto_train_df.reindex(columns=COLUMNS)
jnli_kyoto_train_df.columns = COLUMNS
jnli_kyoto_valid_df = pd.read_csv( jnli_kyoto_valid_filepath , encoding='utf-8' , header=None, names=JNLI_KYOTO_COLUMNS )
jnli_kyoto_valid_df.columns = COLUMNS

jnli_kyoto_train_df.index = len(jglue_train_df) + len(jsick_train_df.index) + jnli_kyoto_train_df.index
jnli_kyoto_valid_df.index = len(jglue_valid_df) + len(jsick_valid_df.index) + jnli_kyoto_valid_df.index

jnli_kyoto_train_df
```

```python
train_df = pd.concat( [ jglue_train_df , jsick_train_df , jnli_kyoto_train_df ] , join='outer' )
valid_df = pd.concat( [ jglue_valid_df , jsick_valid_df , jnli_kyoto_valid_df ] , join='outer' )
print(f'TRAIN\n\tJGLUE: {len(jglue_train_df.index)}\n\tJSICK: {len(jsick_train_df.index)}\n\tJNLI: {len(jnli_kyoto_train_df.index)}\n\tALL: {len(train_df.index)}')
print(f'VALID\n\tJGLUE: {len(jglue_valid_df.index)}\n\tJSICK: {len(jsick_valid_df.index)}\n\tJNLI: {len(jnli_kyoto_valid_df.index)}\n\tALL: {len(valid_df.index)}')

train_filepath = '/content/drive/MyDrive/Colab Notebooks/SeniorThesis/DATASETS/train_dataset.csv'
valid_filepath = '/content/drive/MyDrive/Colab Notebooks/SeniorThesis/DATASETS/valid_dataset.csv'
train_df.to_csv( train_filepath )
valid_df.to_csv( valid_filepath )
```
